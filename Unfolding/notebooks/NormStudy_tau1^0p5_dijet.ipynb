# Functions to refill histograms to new bins

```python
def Fill1D(h0,h,bins):
    N=h0.GetNbinsX()
    content=np.zeros((len(bins)-1),dtype=float)
    for i in range(len(bins)-1):
        content[i]=0
    for i in range(N):
        center=h0.GetBinCenter(i+1)
        index=np.digitize(np.array([center]),bins)
        if index[0] < len(bins):
            content[index[0]-1]+=h0.GetBinContent(i+1)
    for i in range(len(bins)-1):
        h.SetBinContent(i+1,content[i])

def Fill2D(h0,h,binx,biny):
    Nx=h0.GetNbinsX()
    Ny=h0.GetNbinsY()
    hx=h0.ProjectionX()
    hy=h0.ProjectionY()
    content=np.zeros((len(binx)-1,len(biny)-1),dtype=float)
    for i in range(len(binx)-1):
        for j in range(len(biny)-1):
            content[i][j]=0
    for i in range(Nx):
        for j in range(Ny):
            centerx=hx.GetBinCenter(i+1)
            indexx=np.digitize(np.array([centerx]),binx)
            centery=hy.GetBinCenter(j+1)
            indexy=np.digitize(np.array([centery]),biny)
            if indexx[0] < len(binx) and indexy[0] < len(biny):
                content[indexx[0]-1][indexy[0]-1]+=h0.GetBinContent(i+1,j+1)
    for i in range(len(binx)-1):
        for j in range(len(biny)-1):
            h.SetBinContent(i+1,j+1,content[i][j])
```

# Do unfolding

```python
def DoUnfolding0(Response,Data_reco,SysUnc,gen_bins):
    
    tunfolderData = rt.TUnfoldDensity(Response,rt.TUnfold.kHistMapOutputHoriz,rt.TUnfold.kRegModeCurvature, rt.TUnfold.kEConstraintNone, rt.TUnfoldDensity.kDensityModeBinWidth)
    tunfolderData.SetInput(Data_reco)
    for source in ['jesTotal',"jer","pu"]:
        for ud in ['Up','Down']:
            tunfolderData.AddSysError(SysUnc[source+ud],source+ud,rt.TUnfold.kHistMapOutputHoriz,rt.TUnfoldSys.kSysErrModeMatrix)
    
    tunfolderData.DoUnfold(0.)
    
    Data_unfolded=tunfolderData.GetOutput("Data_unfolded")
    
    N=Data_unfolded.GetNbinsX()
    
    cov_total=tunfolderData.GetEmatrixTotal("Total Covariance Matrix")
    cov_uncorr=tunfolderData.GetEmatrixSysUncorr("Covariance Matrix from Uncorrelated Uncertainties")
    cov_input=tunfolderData.GetEmatrixInput("Covariance Matrix from Stat Uncertainties of Input Data")
    
    Unc_stat=Data_unfolded.Clone("Statistical Uncertainties")
    Unc_stat.Reset()
    Unc_total=Data_unfolded.Clone("Statistical and Unfolding Uncertainties")
    Unc_total.Reset()
    Unc_sys={}
    Unc_sys["Total"]=Data_unfolded.Clone("Systematic Uncertainties")
    Unc_sys["Total"].Reset()
    
    for i in range(1,N+1):
        stat1=cov_uncorr.GetBinContent(i,i)
        stat2=cov_input.GetBinContent(i,i)
        total=cov_total.GetBinContent(i,i)
        Unc_stat.SetBinContent(i,np.sqrt(stat1+stat2))
        Unc_total.SetBinContent(i,np.sqrt(total))
    #Unc_stat.Divide(Data_unfolded)
    #Unc_total.Divide(Data_unfolded)
        
    for source in ['jesTotal',"jer","pu"]:
        for ud in ['Up','Down']:
            Unc_sys[source+ud]=tunfolderData.GetDeltaSysSource(source+ud, "1 #sigma Systematic Uncertainties:"+source+ud, "-1#sigma")
            #Unc_sys[source+ud].Divide(Data_unfolded)
        Unc_sys[source+"Total"]=Unc_sys[source+ud].Clone("1 #sigma Systematic Uncertainties:"+source+"Total")
        Unc_sys[source+"Total"].Reset()
        for i in range(1,N+1):
            sys1=Unc_sys[source+"Up"].GetBinContent(i)
            sys2=Unc_sys[source+"Down"].GetBinContent(i)
            Unc_sys[source+"Total"].SetBinContent(i,np.sqrt(sys1**2+sys2**2))
        Unc_sys["Total"].Add(Unc_sys[source+"Total"])
    
    return Data_unfolded,Unc_total,Unc_stat,Unc_sys

def SetStyle(h,color,size,width,style1,style2):
    h.SetLineColor(color)
    h.SetMarkerColor(color)
    h.SetMarkerSize(size)
    h.SetMarkerStyle(style1)
    h.SetFillColor(0)
    h.SetLineWidth(width)
    h.SetLineStyle(style2)

def DoUnfolding(Response,Data_reco,SysUnc,gen_bins):
    Data_unfolded,Unc_total,Unc_stat,Unc_sys=DoUnfolding0(Response,Data_reco,SysUnc,gen_bins)
    Nom=Data_reco.Integral()
    #Nom=Data_unfolded.Integral()
    Data_unfolded.Scale(1./Nom)
    Unc_total.Scale(1./Nom)
    Unc_stat.Scale(1./Nom)
    Unc_sys["Total"].Scale(1./Nom)
    for source in ['jesTotal',"jer","pu"]:
        for ud in ['Up','Down']:
            Unc_sys[source+ud].Scale(1./Nom)
    
    Data_reco.Scale(1./Data_reco.Integral())
    Data_unfolded_,Unc_total_,Unc_stat_,Unc_sys_=DoUnfolding0(Response,Data_reco,SysUnc,gen_bins)
    
    SetStyle(Data_unfolded,rt.kRed,1.5,4,22,1)
    SetStyle(Unc_total,rt.kRed,1.5,4,22,2)
    SetStyle(Unc_stat,rt.kRed,1.5,4,22,3)
    SetStyle(Unc_sys["Total"],rt.kRed,1.5,4,22,4)
    for source in ['jesTotal',"jer","pu"]:
        for ud in ['Up','Down']:
            SetStyle(Unc_sys[source+ud],rt.kRed,1.5,4,22,4)
        SetStyle(Unc_sys[source+"Total"],rt.kRed,1.5,4,22,4)
    
    SetStyle(Data_unfolded_,rt.kBlue,1.5,2,22,1)
    SetStyle(Unc_total_,rt.kBlue,1.5,2,22,2)
    SetStyle(Unc_stat_,rt.kBlue,1.5,2,22,3)
    SetStyle(Unc_sys_["Total"],rt.kBlue,1.5,2,22,4)
    for source in ['jesTotal',"jer","pu"]:
        for ud in ['Up','Down']:
            SetStyle(Unc_sys_[source+ud],rt.kBlue,1.5,2,22,4)
        SetStyle(Unc_sys_[source+"Total"],rt.kBlue,1.5,2,22,4)
    
    hs=rt.THStack("Data_unfolded #tau_{1}^{0.5}","Data_unfolded #tau_{1}^{0.5}")
    Data_unfolded.Scale(1,"width")
    Data_unfolded_.Scale(1,"width")
    hs.Add(Data_unfolded,"Hist")
    hs.Add(Data_unfolded_,"Hist")
    
    leg=rt.TLegend(0.4, 0.65, 0.9, 0.8)
    leg.AddEntry(Data_unfolded,"Data_unfolded w/o nom","f")
    leg.AddEntry(Data_unfolded_,"Data_unfolded","f")
    leg.SetLineColor(0)
    leg.SetBorderSize(0)
    leg.SetFillStyle(0)
    
    hs_unc=rt.THStack("Uncertainties #tau_{1}^{0.5}","Uncertainties #tau_{1}^{0.5}")
    hs_unc.Add(Unc_sys["Total"],"Hist")
    hs_unc.Add(Unc_total,"Hist")
    hs_unc.Add(Unc_stat,"Hist")
    hs_unc.Add(Unc_sys_["Total"],"Hist")
    hs_unc.Add(Unc_total_,"Hist")
    hs_unc.Add(Unc_stat_,"Hist")
    
    leg_unc=rt.TLegend(0.15, 0.15, 0.65, 0.55)
    leg_unc.AddEntry(Unc_sys["Total"],"Systematic unc w/o norm","f")
    leg_unc.AddEntry(Unc_total,"Statistical unc+Unfolding unc w/o norm","f")
    leg_unc.AddEntry(Unc_stat,"Statistical unc w/o norm","f")
    leg_unc.AddEntry(Unc_sys_["Total"],"Systematic unc","f")
    leg_unc.AddEntry(Unc_total_,"Statistical unc+Unfolding unc","f")
    leg_unc.AddEntry(Unc_stat_,"Statistical unc","f")
    leg_unc.SetLineColor(0)
    leg_unc.SetBorderSize(0)
    leg_unc.SetFillStyle(0)
    
    hs_sysunc={}
    leg_sysunc={}
    for source in ['jesTotal',"jer","pu"]:
        for ud in ['Up','Down']:
            hs_sysunc[source+ud]=rt.THStack("1 #sigma shift: "+source+ud,"1 #sigma shift: "+source+ud)
            hs_sysunc[source+ud].Add(Unc_sys[source+ud],"Hist")
            hs_sysunc[source+ud].Add(Unc_sys_[source+ud],"Hist")
            leg_sysunc[source+ud]=rt.TLegend(0.5, 0.5, 0.9, 0.8,source+ud)
            leg_sysunc[source+ud].AddEntry(Unc_sys[source+ud],"Without norm","f")
            leg_sysunc[source+ud].AddEntry(Unc_sys_[source+ud],"With norm","f")
            leg_sysunc[source+ud].SetLineColor(0)
            leg_sysunc[source+ud].SetBorderSize(0)
            leg_sysunc[source+ud].SetFillStyle(0)
    
    return hs,leg,hs_unc,leg_unc,hs_sysunc,leg_sysunc,Data_unfolded,Data_unfolded_
```
